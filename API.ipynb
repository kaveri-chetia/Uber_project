```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
#Reading the data
df =  pd.read_csv("./data/cleaned_data.csv")
columns_to_keep = ['pickup_datetime', 'pickup_longitude','pickup_latitude']
df = df[columns_to_keep]
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

# Round down to the nearest hour
df['rounded_timestamp'] = df['pickup_datetime'].dt.floor('H')
df = df.drop('pickup_datetime', axis=1)
df
```

```python


# Format the datetime column to 'YYYY-MM-DD:HH'
df['rounded_timestamp'] = df['rounded_timestamp'].dt.strftime('%Y-%m-%d:%H')
df
```

```python
df = df.rename(columns={
    'pickup_longitude': 'longitude',
    'pickup_latitude': 'latitude',
    'rounded_timestamp': 'start_time',
})
```

```python
def add_one_hour(start_time):
    # Parse the string into a datetime object
    dt = datetime.strptime(start_time, '%Y-%m-%d:%H')
    # Add one hour
    end_time = dt + timedelta(hours=1)
    # Convert back to 'YYYY-MM-DD:HH' format
    return end_time.strftime('%Y-%m-%d:%H')

# Apply the function to the 'start_time' column to create 'end_time'
df['end_time'] = df['start_time'].apply(add_one_hour)
```
